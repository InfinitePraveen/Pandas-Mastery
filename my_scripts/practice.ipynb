```python
import pandas as pd
```

```python
# Series : A Pandas 1-Dimentional labelled array that can hold any data type. Think of it like a single column in a spreadsheet(1-dimensional)
data = [100, 102, 104]
series = pd.Series(data)
print(series)
```

```python
# Defining the index
data = [100, 102, 104]
series = pd.Series(data, index=["a", "b", "c"]) # By adding index
print(series)
```

```python
# Accessing the index with loc operator
data = [100, 102, 104]
aeries = pd.Series(data)
print(series.loc["a"])
print(series.loc["b"])
print(series.loc["c"])
```

```python
# Changing the value by index
data = [100, 102, 104]

series = pd.Series(data, index=["a", "b", "c"])
series.loc["a"] = 125
series.loc["b"] = 150
series.loc["c"] = 200

print(series)
```

```python
# Locate the value by integer position by .iloc operator
data = [100, 102, 104]

series = pd.Series(data, index=["a", "b", "c"])
print(series.iloc[0])
print(series.iloc[1])
print(series.iloc[2])
```

```python
# Filtering the values
data = [100, 102, 104, 200, 202]

series = pd.Series(data, index=["a", "b", "c", "d", "e"])
print(series[series >= 200])# Greater than or equal to 200
print(series[series < 200])# Less than 200
```

```python
# With dictionary
calories = {"Day 1" : 1750,
            "Day 2" : 2000,
            "Day 3" : 2200}

calories = pd.Series(calories)
print(calories)
```

```python
# Accesing the dictionary key wise
calories = {"Day 1" : 1750, "Day 2" : 2100, "Day 3" : 1700}
calories = pd.Series(calories)
print(calories["Day 1"])
print(calories["Day 2"])
print(calories["Day 3"])
```

```python
# Updating the value in dictonary
calories = {"Day 1" : 1750, "Day 2" : 2100, "Day 3" : 1700}
calories = pd.Series(calories)
# calories["Day 1"] = 2000
# calories["Day 2"] = 2200
# calories["Day 3"] = 2400
calories.loc["Day 1"] = 2000
calories.loc["Day 2"] = 2200
calories.loc["Day 3"] = 2400
print(calories)
```

```python
# Adding the value in dictionary keys
calories = {"Day 1" : 1750, "Day 2" : 2100, "Day 3" : 1700}
calories = pd.Series(calories)
calories["Day 1"] += 500
calories.loc["Day 2"] += 500
calories.iloc[2] += 500
print(calories)
```

```python
# Filtering the values
calories = {"Day 1" : 1750, "Day 2" : 2100, "Day 3" : 1700}
series = pd.Series(calories)
print(series[series >= 2000])
print(series[series < 2000])
```

```python
# Data Frame: A tabular data structure with rows and columns (2-dimensions). Similar to excel spreadsheet.
data = {"Name" : ["Spongebob", "Patrick", "Squidward"],
        "Age" : [30, 35, 50]
        }

data = pd.DataFrame(data, index=["Employee 1", "Employee 2", "Employee 3"])
print(data)
```

```python
# Accessing the dataframe by .loc function
print(data.loc["Employee 1"])
print(data.loc["Employee 2"])
print(data.loc["Employee 3"])
```

```python
# Accessing the dataframe by index
print(data.iloc[0])
print(data.iloc[1])
print(data.iloc[2])
```

```python
# Add a new column
data["Job"] = ["Cook", "N/A", "Cashier"]
print(data)
```

```python
# Add a new row
new_row = pd.DataFrame([{"Name" : "Sandy", "Age" : 28, "Job" : "Engineer"}])
df = pd.concat([data, new_row]) # Index is not set up for this dataframe
print(df)
```

```python
# Adding the index to the value
new_row = pd.DataFrame([{"Name" : "Sandy", "Age" : 28, "Job" : "Engineer"}],
                       index= ["Employee 4"])
df = pd.concat([df, new_row])
print(df)
```

```python
# More practice to add new rows
new_row = pd.DataFrame([{"Name" : "Sandy", "Age" : 28, "Job" : "Engineer"},
                        {"Name" : "Eugne", "Age" : 60, "Job" : "Manager"}],
                       index= ["Employee 4", "Employee 5"])
df = pd.concat([df, new_row]) # Index is not set up for this dataframe
print(df)
```

```python
# Data Cleaning : The process of fixing/removing: incomplete, incorrect, or irrelevant data. ~75% of work done with Pandas is data cleaning

cleaned = df.drop(columns=["Legendary", "No"])
print(cleaned)
```

```python
# 2.Handle missing data
missing_data = df.dropna(subset=["Type2"])
data = df.fillna({"Type2" : "None"})
print(data.to_string())
```

```python
# 3. Fix inconsistent values
df["Type1"] = df["Type1"].replace({"Grass" : "GRASS",
                                   "Fire" : "FIRE",
                                   "Water" : "WATER"})
print(df.to_string())
```

```python
# 4. Standardize text
df = pd.read_csv("data.csv")
df["Name"] = df["Name"].str.lower()
print(df.to_string())
```

```python
# 5. Fix data types
df["Legendary"] = df["Legendary"].astype(bool)
print(df.to_string())
```

```python
# 6. Remove duplicate data
# Have to add duplicate values, at that time I'am not adding the duplicate values.
df = df.drop_duplicates()
print(df.to_string())
```
