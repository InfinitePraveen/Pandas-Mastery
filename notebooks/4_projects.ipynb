## Project 1: Employee Salary Analysis

Using the employees dataset, answer these questions:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
employees = pd.read_csv('../data/employees.csv')
employees
```

```python
# 1. What's the average salary by department?
avg_salary = employees.groupby('department')['salary'].mean()
avg_salary
```

```python
# 2. Who are the top 3 highest paid employees?
top3 = employees.nlargest(3, 'salary')[['name', 'department', 'salary']]
top3
```

```python
# 3. Create a bonus column and visualize
employees['bonus'] = employees['salary'] * 0.10
employees['total'] = employees['salary'] + employees['bonus']

# Bar chart
employees.plot(x='name', y='total', kind='bar')
plt.title('Total Compensation by Employee')
plt.xlabel('Employee')
plt.ylabel('Total ($)')
plt.xticks(rotation=45)
plt.show()
```

## Project 2: Sales Performance Dashboard

Create a simple sales dashboard

```python
# Load sales data
sales = pd.read_csv('../data/sales.csv')
sales['total'] = sales['quantity'] * sales['price']

# Calculate KPIs
total_revenue = sales['total'].sum()
avg_order = sales['total'].mean()
total_items = sales['quantity'].sum()

print(f"📊 Sales Dashboard")
print("=" * 30)
print(f"Total Revenue: ${total_revenue:,.0f}")
print(f"Average Order: ${avg_order:.2f}")
print(f"Total Items Sold: {total_items}")
print(f"Number of Orders: {len(sales)}")
```

```python
# Visualize results
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Sales by product
sales.groupby('product')['total'].sum().plot(kind='bar', ax=axes[0])
axes[0].set_title('Sales by Product')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Revenue ($)')

# Sales by region
sales.groupby('region')['total'].sum().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
axes[1].set_title('Sales by Region')

plt.tight_layout()
plt.show()
```
