# Notebook 1: Getting Started with Pandas

Let's learn the basics of pandas!

## 1. Import pandas

```python
import pandas as pd
import numpy as np
```

## 2. Load Data

```python
# Load employee data
df = pd.read_csv('../data/employees.csv')
df
```

## 3. Look at your data

```python
# First 5 rows
df.head()
```

```python
# Last 5 rows
df.tail()
```

```python
# Basic info
df.info()
```

```python
# Summary statistics
df.describe()
```

## 4. Select columns

```python
# Pick one column
df['name']
```

```python
# Pick multiple columns
df[['name', 'salary']]
```

## 5. Filter rows

```python
# IT department only
df[df['department'] == 'IT']
```

```python
# Salary > 60000
df[df['salary'] > 60000]
```

```python
# Multiple conditions
df[(df['department'] == 'Sales') & (df['salary'] > 60000)]
```

## 6. Add new columns

```python
# Add bonus column (10% of salary)
df['bonus'] = df['salary'] * 0.10
df
```

```python
# Add total compensation
df['total'] = df['salary'] + df['bonus']
df
```
