# Notebook 2: Cleaning Data

Real data is messy. Let's learn how to clean it!

```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('../data/sales.csv')
df
```

## 1. Find Missing Values

```python
# Check for missing values
df.isnull().sum()
```

## 2. Handle Missing Values

```python
# Option 1: Drop rows with missing values
df_clean = df.dropna()
print(f"Original: {len(df)} rows")
print(f"After dropping: {len(df_clean)} rows")
```

```python
# Option 2: Fill missing values
df_filled = df.fillna(0)  # Fill with 0
df_filled
```

## 3. Find Duplicates

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicates
df_unique = df.drop_duplicates()
print(f"Rows after removing duplicates: {len(df_unique)}")
```

## 4. Fix Data Types

```python
# Check data types
df.dtypes
```

```python
# Convert date column
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

## 5. Clean Text Data

```python
# Make text lowercase
df['product'] = df['product'].str.lower()
df
```

```python
# Remove spaces
df['product'] = df['product'].str.strip()
df
```
